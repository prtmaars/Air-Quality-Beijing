## Menentukan Pertanyaan Bisnis

Dataset yang digunakan merupakan dataset kualitas udara dari 12 kota di China, yaitu Aotizhongxin, Changping, Dingling, Dongsi, Guanyuan, Gucheng, Huairou, Nongzhanguan, Shunyi, Tiantan, Wanliu dan Wanshouxigong. Data diambil setiap interval 1 jam dari 1 Maret 2013 sampai dengan 28 Februari 2017. Dataset terdiri dari 17 kolom data dengan rincian sebagai berikut:

1. year: Tahun diambilnya data.
2. month: Bulan diambilnya data.
3. date: Tanggal diambilnya data.
4. hour: Jam diambilnya data.
5. PM2.5: Particulate Matter 2.5, Konsentrasi partikel halus di udara yang berukuran ≤2.5 mikrometer (μg/m³).
6. PM10: Particulate Matter 10, Konsentrasi partikel kasar di udara yang berukuran ≤10 mikrometer (μg/m³).
7. SO2: Sulfur Dioxide, Konsentrasi sulfur dioksida di udara (μg/m³).
8. NO2: Nitrogen Dioxide, Konsentrasi nitrogen dioksida di udara (μg/m³).
9. CO: Carbon Monoxide, Konsentrasi karbon monoksida di udara (μg/m³).
10. O3: Ozone, Konsentrasi ozon di udara (μg/m³).
11. TEMP: Temperature, Suhu udara dalam derajat Celcius (°C).
12. PRES: Pressure, Tekanan atmosfer dalam hPa.
13. DEWP: Dew Point, Titik embun dalam derajat Celcius (°C).
14. RAIN: Rainfall, Curah hujan atau intensitas hujan dalam mm.
15. wd: Wind Direction, Arah angin (dalam singkatan, seperti WNW untuk West-Northwest).
16. WSPM: Wind Speed per Minute, Kecepatan angin dalam m/s.
17. station: Nama stasiun tempat data diambil.

### Cleaning Data

```python
for name, data in datasets:
  print(f"Dataset {name} - Duplikasi: {data.duplicated().sum()}")
```

**Penjelasan:** Dilakukan pengecekan terhadap duplikasi data. Tidak ditemukan duplikasi data pada seluruh data.

```python
missval1 = pd.DataFrame()
for name, data in datasets:
    missval1[name] = data.isna().sum()
print("Missing Value:")
print(missval1)
```

**Penjelasan:** Dilakukan pengecekan terhadap missing-value atau nilai yang hilang pada setiap data. Ditemukan missing-value pada 1 kolom kategorikal dengan nama 'wd' dan 11 kolom numerikal.

```python
for name, data in datasets:
    data["PRES"] = data["PRES"].fillna(value=data["PRES"].mean())
    data["PM2.5"] = data["PM2.5"].fillna(value=data["PM2.5"].mean())
    data["PM10"] = data["PM10"].fillna(value=data["PM10"].mean())
    data["SO2"] = data["SO2"].fillna(value=data["SO2"].mean())
    data["NO2"] = data["NO2"].fillna(value=data["NO2"].mean())
    data["CO"] = data["CO"].fillna(value=data["CO"].mean())
    data["O3"] = data["O3"].fillna(value=data["O3"].mean())
    data["TEMP"] = data["TEMP"].fillna(value=data["TEMP"].mean())
    data["DEWP"] = data["DEWP"].fillna(value=data["DEWP"].mean())
    data["RAIN"] = data["RAIN"].fillna(value=data["RAIN"].mean())
    data["WSPM"] = data["WSPM"].fillna(value=data["WSPM"].mean())
```

**Penjelasan:** Data pada kolom numerikal kemudian diisi dengan nilai rata-rata dari setiap kolomnya.

**Penjelasan:** Dilakukan pengecekan terhadap missing-value yang ada pada kolom kategorikal. Hasilnya didapat bahwa missing-value terbagi menjadi 2, yaitu ketika WSPM = 0 dan WSPM > 0.

**Penjelasan:** Dilakukan pengecekan terhadap nilai-nilai yang terdapat pada kolom 'wd'. Kolom 'wd' merupakan arah mata angin.

```python
for name, data in datasets:
    data.loc[data['wd'].isna() & (data['WSPM'] > 0), 'wd'] = 'unknown'
    data.loc[data['wd'].isna() & (data['WSPM'] == 0), 'wd'] = 'no wind'
```

```python
for i in range(1, 13):
    data_name = f"data_{i}"
    data = globals()[data_name]
    for period in range(1, 5):
        period_data = data[data['PERIOD'] == period]
        for month_offset in range(1, 13):
            actual_month = (month_offset + 1) % 12 + 1
            month_data = period_data[period_data['month'] == actual_month]
            for day in range(1, 32):
                day_data = month_data[month_data['day'] == day]
                if not day_data.empty:
                    var_name = f"{data_name}_period{period}_month{month_offset}_day{day}"
                    globals()[var_name] = day_data
```

**Penjelasan:** Kode diatas digunakan untuk membuat variabel baru dengan format nama 'data_[i]_period[i]_month[i]_day[i]' untuk membagi tiap data berdasarkan hari di tiap bulan di tiap periodnya. Hal ini dilakukan untuk memudahkan analisis data berdasarkan hari di tiap bulan di tiap periodenya.

## Visualization & Explanatory Analysis

```python
dataset_city = [
    ("Aotizhongxin", city_1),
    ("Changping", city_2),
    ("Dingling", city_3),
    ("Dongsi", city_4),
    ("Guanyuan", city_5),
    ("Gucheng", city_6),
    ("Huairou", city_7),
    ("Nongzhanguan", city_8),
    ("Shunyi", city_9),
    ("Tiantan", city_10),
    ("Wanliu", city_11),
    ("Wanshouxigong", city_12)
]
```

```python
cmap = plt.get_cmap("viridis")
city_names = [city for city, _ in dataset_city]
x = range(len(city_names))
```

**Penjelasan:** Kode di atas digunakan untuk mempersiapkan data dan elemen-elemen yang diperlukan untuk membuat visualisasi data.

```python
pm25_values = []
pm10_values = []
so2_values = []
no2_values = []
co_values = []
o3_values = []
temp_values = []
pres_values = []
dewp_values = []
rain_values = []
wspm_values = []
```

**Penjelasan:** Kode di atas merupakan deklarasi dari beberapa list kosong.

```python
for city, data in dataset_city:
    pm25_values.append(data.loc['mean', 'PM2.5'])
    pm10_values.append(data.loc['mean', 'PM10'])
    so2_values.append(data.loc['mean', 'SO2'])
    no2_values.append(data.loc['mean', 'NO2'])
    co_values.append(data.loc['mean', 'CO'])
    o3_values.append(data.loc['mean', 'O3'])
    temp_values.append(data.loc['mean', 'TEMP'])
    pres_values.append(data.loc['mean', 'PRES'])
    dewp_values.append(data.loc['mean', 'DEWP'])
    rain_values.append(data.loc['mean', 'RAIN'])
    wspm_values.append(data.loc['mean', 'WSPM'])
```

**Penjelasan:** Kode di atas digunakan untuk melakukan iterasi pada dataset_city, yang berisi pasangan nama kota (city) dan data kualitas udara (data) dari masing-masing kota. Pada setiap iterasi, kode ini mengekstrak nilai rata-rata (mean) dari berbagai parameter kualitas udara dan cuaca, kemudian menyimpannya ke dalam list yang sesuai.

### Pertanyaan 1: Analisis Kualitas Udara: Kota apa yang memiliki udara paling bersih dan paling kotor dari segi partikel udara?

```python
sorted_pm10 = sorted(zip(pm10_values, pm25_values, city_names), reverse=True, key=lambda x: x[0])
sorted_pm10_values, sorted_pm25_values, sorted_city_names = zip(*sorted_pm10)
width = 0.40
x = range(len(sorted_city_names))
fig, ax = plt.subplots(figsize=(12, 8))
bar1 = ax.bar(x, sorted_pm25_values, width, label='PM2.5', color=cmap(0.5))
bar2 = ax.bar([p + width for p in x], sorted_pm10_values, width, label='PM10', color=cmap(0.8))
ax.set_xlabel('Kota', fontsize=12)
ax.set_ylabel('Konsentrasi (µg/m³)', fontsize=12)
ax.set_title('Perbandingan Rata-Rata PM2.5 dan PM10 dari Setiap Kota', fontsize=14)
ax.set_xticks([p + width / 2 for p in x])
ax.set_xticklabels(sorted_city_names, rotation=45, ha="right")
y_min = 0
y_max = int(max(sorted_pm10_values + sorted_pm25_values) + 20)
ax.set_ylim(y_min, y_max)
ax.set_yticks(range(y_min, y_max + 1, 10))
ax.legend()
plt.tight_layout()
plt.show()
```

**Penjelasan:** PM2.5 atau Particulate Matter 2.5 merupakan konsentrasi partikel halus di udara yang berukuran ≤2.5 mikrometer (μg/m³), sedangkan PM10 atau Particulate Matter 10 merupakan konsentrasi partikel kasar di udara yang berukuran ≤10 mikrometer (μg/m³). Pada dasarnya, perhitungan pada PM10 sudah mencakup dari perhitungan PM2.5. Dalam hal ini, dapat disimpulkan bahwa kota yang memiliki jumlah PM10 paling banyak merupakan kota dengan kualitas udara paling kotor dari segi partiker udara dan kota yang memiliki jumlah PM10 paling sedikit merupakan kota dengan kualitas udara paling bersih dari segi partikel udara. Maka, dalam rentang waktu 1 Maret 2013 sampai dengan 28 Februari 2017, Kota Dingling merupakan kota paling bersih dan Kota Gucheng merupakan kota paling kotor pada kualitas udaranya dari segi partikel udara.

### Pertanyaan 2: Analisis Meteorologi: Kota apa yang memiliki curah hujan paling tinggi dan paling rendah? Bagaimana pengaruh hujan terhadap polusi udara?

```python
sorted_rain = sorted(zip(so2_values, city_names), reverse=True, key=lambda x: x[0])
sorted_rain_values, sorted_city_names = zip(*sorted_rain)
width = 0.75
fig, ax = plt.subplots(figsize=(12, 8))
x = range(len(sorted_city_names))
bar1 = ax.bar(x, sorted_rain_values, width, label='RAIN', color=cmap(0.5))
ax.set_xlabel('Kota', fontsize=12)
ax.set_ylabel('Nilai', fontsize=12)
ax.set_title('Nilai RAIN dari Setiap Kota', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(sorted_city_names, rotation=45, ha="right")
y_min = 0
y_max = int(max(sorted_rain_values) + 5)
ax.set_ylim(y_min, y_max)
ax.set_yticks(range(y_min, y_max + 1, 1))
ax.legend()
plt.tight_layout()
plt.show()
```

Jika dibandingkan dengan grafik pada partikel udara, nampaknya bahwa curah hujan yang tinggi tidak terlalu berpengaruh secara signifikan terhadap kualitas udara. Terlihat bahwa kota dengan kualitas udara paling bersih dari segi partikel udara yaitu Kota Dingling berada di urutan paling terakhir dengan curah hujan paling tinggi, sedangkan Kota Gucheng yang merupakan kota dengan kualitas udara paling kotor dari segi partikel udara berada di urutan ke-7. **Penjelasan:** Dilakukan visualisasi berdasarkan data PM10 atau konsentrasi partikel halus di udara yang berukuran ≤10 mikrometer (μg/m³). Berdasarkan WHO (World Health Organization), konsentrasi PM10 di udara dibagi menjadi 5 kategori yaitu:
- Baik 0–50 μg/m³
- Sedang	50–100 μg/m³
- Tidak Sehat	100–250 μg/m³
- Sangat Tidak Sehat 250–350 μg/m³
- Berbahaya >350 μg/m³

Berdasarkan grafik di atas, 8 dari 12 kota berada pada kategori Tidak Sehat, sedangkan 4 lainnya berada pada kategori Sedang.

```python
category_limits_pm25 = [15, 35, 75, 150, 300]
category_limits_pm10 = [50, 100, 250, 350, 500]
city_status = []
def calculate_aqi(pm_value, category_limits):
    if pm_value <= category_limits[0]:
        return (pm_value / category_limits[0]) * 50
    elif pm_value <= category_limits[1]:
        return ((pm_value - category_limits[0]) / (category_limits[1] - category_limits[0])) * 50 + 51
    elif pm_value <= category_limits[2]:
        return ((pm_value - category_limits[1]) / (category_limits[2] - category_limits[1])) * 50 + 101
    elif pm_value <= category_limits[3]:
        return ((pm_value - category_limits[2]) / (category_limits[3] - category_limits[2])) * 50 + 151
    elif pm_value <= category_limits[4]:
        return ((pm_value - category_limits[3]) / (category_limits[4] - category_limits[3])) * 50 + 201
    else:
        return 301
```

**Penjelasan:** Kode di atas digunakan untuk mengkategorikan kota berdasarkan dengan AQI (Air Quality Index).

```python
categories = {'Baik': [], 'Sedang': [], 'Tidak Sehat': [], 'Sangat Tidak Sehat': [], 'Berbahaya': []}
for city, data in dataset_city:
    pm25_mean = data.loc['mean', 'PM2.5']
    pm10_mean = data.loc['mean', 'PM10']
    aqi_pm25 = calculate_aqi(pm25_mean, category_limits_pm25)
    aqi_pm10 = calculate_aqi(pm10_mean, category_limits_pm10)
    aqi = max(aqi_pm25, aqi_pm10)
    if aqi <= 50:
        status = 'Good'
        categories['Baik'].append(city)
    elif aqi <= 100:
        status = 'Moderate'
        categories['Sedang'].append(city)
    elif aqi <= 150:
        status = 'Unhealthy'
        categories['Tidak Sehat'].append(city)
    elif aqi <= 200:
        status = 'Very Unhealthy'
        categories['Sangat Tidak Sehat'].append(city)
    else:
        status = 'Hazardous'
        categories['Berbahaya'].append(city)
    city_status.append({'city': city, 'air_quality_status': status})
    print(f"{city}: PM2.5 = {pm25_mean}, PM10 = {pm10_mean}, AQI2.5 = {aqi_pm25}, AQI10 = {aqi_pm10}, AQI = {aqi}")
```

**Penjelasan:** Dilakukan perhitungan AQI (Air Quality Index) untuk setiap kotanya.
